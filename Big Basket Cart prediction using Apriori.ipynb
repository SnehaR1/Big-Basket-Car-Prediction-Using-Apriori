```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/Dataset_master.xlsx - Big Basket.com Cart Apriori (Pr.csv',header=None)
data.head()
```

```python
data.shape
```

```python
transactions=[]
for i in range(0,7219):
  transactions.append([str(data.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
rules=apriori(transactions=transactions,min_confidence=0.2,min_support=0.003,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(rules)
print(results)
```

```python
def inspect(results):
    product1         = [tuple(result[2][0][0])[0] for result in results]
    product2         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(product1, product2, supports, confidences, lifts))
DataFrame_intelligence = pd.DataFrame(inspect(results), columns = ['product1', 'product1', 'Support', 'Confidence', 'Lift'])
```
